```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("datas/top5-players.csv") 

print(df.columns)
```

```python
# Colonnes numériques à analyser
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

# Statistiques par club (Squad)
stats_par_club = df.groupby('Squad')[numerical_cols].agg(['mean', 'median', 'std'])
display(stats_par_club)
```
