```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df = pd.read_csv("datas/top5-players.csv")

print("DESCRIPTION INITIALE")
print("Shape :", df.shape)
print("\nTypes de données :")
print(df.dtypes)
print("\nAperçu :")
display(df.head())  

print("\nDOUBLONS")
nb_doublons = df.duplicated().sum()
print("Nombre de doublons :", nb_doublons)

print("\nVALEURS MANQUANTES")
missing_report = pd.DataFrame({
    'Valeurs manquantes': df.isnull().sum(),
    '% manquant': (df.isnull().mean() * 100).round(2)
})
display(missing_report)
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True, bins=10, color='skyblue')
    plt.title(f'Distribution de {col}')
    plt.xlabel(col)
    plt.ylabel('Fréquence')
    plt.show()


for col in numerical_cols:
    moyenne = np.mean(df[col])
    med = np.median(df[col])
    ecart_type = np.std(df[col])
    print(f"\nStatistiques pour {col}:")
    print(f"  Moyenne : {moyenne:.2f}")
    print(f"  Médiane : {med:.2f}")
    print(f"  Écart-type : {ecart_type:.2f}")

```

```python
def clean_csv(path):
    df = pd.read_csv(path)
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    df.columns = df.columns.str.strip().str.lower()
    return df

```

```python
df1 = clean_csv("datas/top5-players.csv")
df2 = clean_csv("Defensive.csv")
df3 = clean_csv("keepers.csv")
df4 = clean_csv("Passing.csv")
```

```python

print("df1 :", df1.columns.tolist())
print("df2 :", df2.columns.tolist())
print("df3 :", df3.columns.tolist())
print("df4 :", df4.columns.tolist())
```

```python
df_total = pd.concat([df1, df2, df3, df4], axis=1)
display(df_total.head())
```

```python
if 'player' in df_total.columns:
    print("\nDOUBLONS PAR NOM DE JOUEUR")
    nb_doublons_nom = df_total.duplicated(subset=['player']).sum()
    print("Nombre de doublons par nom :", nb_doublons_nom)

    doublons_joueurs = df_total[df_total.duplicated(subset=['player'], keep=False)]
    if not doublons_joueurs.empty:
        print("Joueurs en doublon :")
        display(doublons_joueurs)
    else:
        print("Aucun doublon trouvé.")
else:
    print("La colonne 'player' n'existe pas dans ce CSV.")

print("\nVALEURS MANQUANTES")
missing_report = pd.DataFrame({
    'Valeurs manquantes': df_total.isnull().sum(),
    '% manquant': (df_total.isnull().mean() * 100).round(2)
})
display(missing_report)

numerical_cols = df_total.select_dtypes(include=np.number).columns.tolist()



for col in numerical_cols:
    moyenne = np.mean(df_total[col])
    med = np.median(df_total[col])
    ecart_type = np.std(df_total[col])
    print(f"\nStatistiques pour {col}:")
    print(f"  Moyenne : {moyenne:.2f}")
  
if 'squad' in df_total.columns:
    stats_par_club = df_total.groupby('squad')[numerical_cols].count()
    print("\nNombre de valeurs par colonne pour chaque club :")
    display(stats_par_club)
    if 'gls' in df_total.columns:
        plt.figure(figsize=(8,4))
        sns.boxplot(x='squad', y='gls', data=df_total)
        plt.title("Distribution des buts par club")
        plt.xlabel("Club")
        plt.ylabel("Buts")
        plt.show()
```
