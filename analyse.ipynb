```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df = pd.read_csv("datas/top5-players.csv")  
df.columns = df.columns.str.strip().str.lower()
print("Colonnes disponibles :", df.columns.tolist())

print("\nDESCRIPTION INITIALE")
print("Shape :", df.shape)
print("\nTypes de données :")
print(df.dtypes)
print("\nAperçu :")
display(df.head())  


if 'player' in df.columns:
    print("\nDOUBLONS PAR NOM DE JOUEUR")
    nb_doublons_nom = df.duplicated(subset=['player']).sum()
    print("Nombre de doublons par nom :", nb_doublons_nom)
    doublons_joueurs = df[df.duplicated(subset=['player'], keep=False)]
    if not doublons_joueurs.empty:
        print("Joueurs en doublon :")
        display(doublons_joueurs)
    else:
        print("Aucun doublon trouvé.")
print("\nVALEURS MANQUANTES")
missing_report = pd.DataFrame({
    'Valeurs manquantes': df.isnull().sum(),
    '% manquant': (df.isnull().mean() * 100).round(2)
})
display(missing_report)

numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True, bins=10, color='skyblue')
    plt.title(f'Distribution de {col}')
    plt.xlabel(col)
    plt.ylabel('Fréquence')
    plt.show()

for col in numerical_cols:
    moyenne = np.mean(df[col])
    med = np.median(df[col])
    ecart_type = np.std(df[col])
    print(f"\nStatistiques pour {col}:")
    print(f"  Moyenne : {moyenne:.2f}")
 
```
